```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})

print("=== 1. 按 gender 分组，score 的均值和最大值 ===")
display(df.groupby('gender')['score'].agg(['mean', 'max']))

print("\n=== 2. 按 gender 分组，统计每组人数 ===")
display(df.groupby('gender').size())
```

```python
# 升序排序（默认 ascending=True，可以省略）
df_sorted = df.sort_values(by='score', ascending=True)
df_sorted
```

```python
# 添加一行重复数据（Tom, 85, 男）
new_row = pd.DataFrame([{'name': 'Tom', 'score': 85, 'gender': '男'}])
df_with_dup = pd.concat([df, new_row], ignore_index=True)

print("添加重复行后：")
display(df_with_dup)

# 去重（默认保留第一次出现的行）
df_deduped = df_with_dup.drop_duplicates()

print("\n去重后：")
display(df_deduped)
```

```python
import pandas as pd

# ========== 1. 横向合并 ==========
df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})
print("=== 横向合并 ===")
print("df1：")
display(df1)
print("df2：")
display(df2)
df_merged = pd.merge(df1, df2, on="id")
print("合并后：")
display(df_merged)

# ========== 2. 纵向拼接 ==========
df_a = pd.DataFrame({"val": [1, 2]})
df_b = pd.DataFrame({"val": [3, 4]})
print("\n=== 纵向拼接 ===")
print("df_a：")
display(df_a)
print("df_b：")
display(df_b)
df_concat = pd.concat([df_a, df_b], axis=0, ignore_index=True)
print("拼接后：")
display(df_concat)
```

```python
import pandas as pd

# 1. 创建并转换时间列
dates_str = ["2025-01-01", "2025-01-02", "2025-01-03"]
df = pd.DataFrame({"date": dates_str})
df["date"] = pd.to_datetime(df["date"])

print("=== 转换后数据 ===")
display(df)
print("date 列数据类型：", df["date"].dtype)

# 2. 提取月份、星期
df["month"] = df["date"].dt.month
df["weekday_num"] = df["date"].dt.weekday
df["weekday_name"] = df["date"].dt.day_name()

print("\n=== 提取月份和星期后 ===")
display(df)
```

```python
import pandas as pd
import numpy as np

# 1. 原始数据
df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})
print("=== 原始数据 ===")
display(df)

# 2. 标准化 score
mean_score = df['score'].mean()
std_score = df['score'].std()
df['score_std'] = (df['score'] - mean_score) / std_score
print("\n=== 标准化后数据 ===")
display(df)

# 3. 透视表（按 gender 求 score 中位数）
pivot = df.pivot_table(index='gender', values='score', aggfunc='median')
print("\n=== 透视表（gender × score 中位数） ===")
display(pivot)
```
